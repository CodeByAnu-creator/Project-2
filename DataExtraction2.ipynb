```python
#As raw data file has >1 sheet of data, merging is essential for Vehicle image column
#Merging of 2 sheets in .excel file to fill the data of vehicle images column
import pandas as pd
import numpy as np # Make sure numpy is imported for NaN handling

# --- Step 1: Load the main ride data from the 'July' sheet ---
# This remains the same.
df = pd.read_excel("C:/Users/abhij/OneDrive/Documents/Anu_Guvi/Project2/OLA_DataSet.xlsx", sheet_name='July')

print("\n--- Initial Load Check for 'July' sheet ---")
print(df.info())
print("\nSample from 'July' sheet (check Vehicle Images column here - it should still be empty for now):")
print(df[['Booking_ID', 'Vehicle_Type', 'Vehicle Images']].head())


# --- Step 2: Load the vehicle image links from 'Sheet2' (without a header) ---
# Use header=None to tell pandas that the first row is not a header.
df_images = pd.read_excel("C:/Users/abhij/OneDrive/Documents/Anu_Guvi/Project2/OLA_DataSet.xlsx", sheet_name='Sheet2', header=None)

# IMPORTANT: Inspect df_images.head() to confirm column indices (0, 1, etc.)
# If your Vehicle Type is NOT in column 0, and Image Link is NOT in column 1,
# you MUST adjust the rename and merge parts below.
print("\n--- Initial Load Check for 'Sheet2' (Vehicle Images) ---")
print(df_images.info())
print("\nSample from 'Sheet2' (CHECK COLUMN INDICES HERE):")
print(df_images.head())


# --- Step 3: Rename columns in df_images and Merge into the main DataFrame (df) ---
# Rename the columns to something meaningful based on your verification in Step 2.
# Assuming column 0 is 'Vehicle_Type' and column 1 is 'Vehicle Images' (the link itself).
df_images.rename(columns={0: 'Vehicle_Type', 1: 'Vehicle Images'}, inplace=True)

# Drop duplicate rows in df_images if a Vehicle_Type appears multiple times with different links
# (e.g., if 'Sheet2' has multiple links for 'Mini', decide which one to keep, or average/list them)
# For simplicity, we'll assume each Vehicle_Type has a unique primary image.
df_images.drop_duplicates(subset=['Vehicle_Type'], inplace=True)


# The 'Vehicle Images' column in df (from 'July' sheet) is currently empty/NaN.
# We will drop it and then merge the correct 'Vehicle Images' from df_images.
df.drop(columns=['Vehicle Images'], inplace=True, errors='ignore')

# Perform the merge. We'll left merge to keep all original ride data.
df = pd.merge(df, df_images[['Vehicle_Type', 'Vehicle Images']], on='Vehicle_Type', how='left')


print("\n--- DataFrame Info After Merging Vehicle Images ---")
print(df.info())
print("\nSample of Vehicle Images column after merge (should now have links):")
print(df[['Booking_ID', 'Vehicle_Type', 'Vehicle Images']].head())

# --- NOW, YOU CAN PASTE THE ENTIRE CONSOLIDATED DATA CLEANING & FEATURE ENGINEERING BLOCK HERE ---
# (The one I provided in the previous turn, including all fillna, timestamp creation,
# boolean flags, and category conversions.)

# After the consolidated cleaning block, re-run your final checks:
# print("\n--- Final df.isnull().sum() after full cleaning ---")
# print(df.isnull().sum())

# print("\n--- Final df.info() after full cleaning ---")
# print(df.info())
```

```python
# --- CONSOLIDATED DATA CLEANING & FEATURE ENGINEERING ---

# 1. Handle Missing Values in Categorical Columns
# Fill NaN with specific placeholder strings based on context
df['Canceled_Rides_by_Customer'].fillna('Not Canceled by Customer', inplace=True)
df['Canceled_Rides_by_Driver'].fillna('Not Canceled by Driver', inplace=True)
df['Incomplete_Rides_Reason'].fillna('Not Incomplete', inplace=True)
df['Payment_Method'].fillna('Unknown', inplace=True)
# For 'Incomplete_Rides', fill NaN with 'No' as per previous discussion, assuming 0/No means not incomplete
df['Incomplete_Rides'].fillna('No', inplace=True)


# 2. Handle Missing Values in Numeric Columns
# Convert to numeric, leaving NaN if present, or fill with 0 if that's desired for calculations
df['V_TAT'] = pd.to_numeric(df['V_TAT'], errors='coerce')
df['C_TAT'] = pd.to_numeric(df['C_TAT'], errors='coerce')
df['Driver_Ratings'] = pd.to_numeric(df['Driver_Ratings'], errors='coerce')
df['Customer_Rating'] = pd.to_numeric(df['Customer_Rating'], errors='coerce')

# Optional: If you want to fill these NaNs with 0 instead of leaving them as NaN
# df['V_TAT'].fillna(0, inplace=True)
# df['C_TAT'].fillna(0, inplace=True)
# df['Driver_Ratings'].fillna(0, inplace=True)
# df['Customer_Rating'].fillna(0, inplace=True)


# 3. Convert and combine Date and Time to a single Timestamp column
# Ensure 'Date' column is in datetime format
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
# Convert 'Time' column to string format for concatenation
df['Time_Str'] = df['Time'].astype(str)
# Combine Date and Time as strings and convert to datetime
df['Timestamp'] = pd.to_datetime(df['Date'].dt.strftime('%Y-%m-%d') + ' ' + df['Time_Str'], errors='coerce')

# Drop original 'Date', 'Time', 'Time_Str' columns and 'Time_Delta' if it was created temporarily
# Ensure Time_Delta is dropped if it somehow persists as a temporary column from previous attempts
df = df.drop(columns=['Date', 'Time', 'Time_Str'], errors='ignore') # 'errors=ignore' prevents error if column already dropped
if 'Time_Delta' in df.columns:
    df = df.drop(columns=['Time_Delta'])


# 4. Extract temporal features from 'Timestamp'
df['Hour_of_Day'] = df['Timestamp'].dt.hour
df['Day_of_Week'] = df['Timestamp'].dt.day_name()
df['Month'] = df['Timestamp'].dt.month_name()
df['Day_of_Month'] = df['Timestamp'].dt.day
df['Weekday_Weekend'] = df['Timestamp'].dt.weekday.apply(lambda x: 'Weekend' if x >= 5 else 'Weekday')


# 5. Create CORRECTED boolean flags for ride status
# Base these directly on Booking_Status for clarity and accuracy
df['Is_Canceled'] = (df['Booking_Status'].isin(['Canceled by Customer', 'Canceled by Driver', 'Driver Not Found']))
df['Is_Customer_Canceled'] = (df['Booking_Status'] == 'Canceled by Customer')
df['Is_Driver_Canceled'] = (df['Booking_Status'] == 'Canceled by Driver')
df['Is_Incomplete'] = (df['Incomplete_Rides'] == 'Yes') # Assuming 'Yes' in Incomplete_Rides means incomplete


# 6. Convert relevant object columns to 'category' dtype for memory efficiency
categorical_cols = [
    'Booking_Status', 'Vehicle_Type', 'Pickup_Location', 'Drop_Location',
    'Canceled_Rides_by_Customer', 'Canceled_Rides_by_Driver', 'Incomplete_Rides',
    'Incomplete_Rides_Reason', 'Payment_Method', 'Day_of_Week', 'Month',
    'Weekday_Weekend'
]

for col in categorical_cols:
    if col in df.columns:
        df[col] = df[col].astype('category')

# --- END OF CONSOLIDATED DATA CLEANING & FEATURE ENGINEERING ---

# After this block, you can then add your df.info(), df.head() and df.to_csv() calls
print("\n--- DataFrame Info After Re-Cleaning & Feature Engineering ---")
df.info()

print("\n--- First 5 Rows After Re-Cleaning & Feature Engineering ---")
print(df.head())
```
